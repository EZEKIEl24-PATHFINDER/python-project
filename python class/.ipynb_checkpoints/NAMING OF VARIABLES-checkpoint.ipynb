```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Line plot on x,y samples
plt.plot(x,y, marker = '*', color = "green")
plt.title("Sample plot for X,Y Variables")
plt.xlabel("X_Variables")
plt.ylabel("Y_Variables")
plt.grid(True)
plt.show()
```
